````python
class Horse:
    def __init__(self, species, name, age, color):
     self.species = species
     self.name = name
     self.age = age
     self.color = color
    def greetings(self):
        print("Hello, my name is ", self.name)
    def __str__(self):
        return f"{self.name} is a {self.color} {self.species}"

bucefalo = Horse("horse", "Bucefalo", 3, "brown")
bucefalo.greetings()
````
